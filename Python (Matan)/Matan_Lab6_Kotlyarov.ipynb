```python
import numpy as np
import matplotlib.pyplot as plt

def plot_func(f, g, a, b, n):
    
    x = np.linspace(a, b, n)
    y1 = f(x)
    y2 = g(x)

    plt.plot(x, y1, label="y=f(x)")
    plt.plot(x, y2, label="y=g(x)")

    #    
    plt.fill_between(x, y1, y2, where=y1 <= y2, interpolate=True, hatch='||', alpha=0.5)

    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')

    plt.show()
    
def f(x):
    return x**2

def g(x):
    return 5*np.sin(x)

a, b = 0, 4
n = 100

plot_func(f, g, a, b, n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_func(f, a, b, n):
    x = np.linspace(a, b, n)
    y = np.linspace(a, b, n)
    x, y = np.meshgrid(x, y)
    z = f(x, y)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    ax.plot_surface(x, y, z, cmap="coolwarm")

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')

    plt.show()

def plot_func_contour(f, g, a, b, n):
 
    x = np.linspace(a, b, n)
    y = np.linspace(a, b, n)
    x, y = np.meshgrid(x, y)
    z1 = f(x, y)
    z2 = g(x, y)

    fig, ax = plt.subplots()

    # Plot the contours of the region defined by f(x,y)<g(x,y)
    c = ax.contourf(x, y, z2 < z1, alpha=0.5)

    ax.axhline(y=1, color='black')
    ax.set_xlim(a, b)
    ax.set_ylim(a, b)

    fig.colorbar(c)

    plt.show()
```
