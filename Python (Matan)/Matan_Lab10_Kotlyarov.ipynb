```python
import numpy as np
import sympy as sp
from sympy import *
import matplotlib.pyplot as plt

x = np.linspace(-2.5, 2.5, 100)
y1 = 2*x
y2 =x**2

plt.plot(x, y1, 'r-', label='2x')
plt.plot(x, y2, 'b-', label='x^2')

plt.xlabel('x')
plt.ylabel('y(x)')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.fill_between(x,y1,y2,where=(y1>y2),color='g',alpha=0.5)
plt.legend()
plt.grid(True)
plt.axis('equal')
```

```python
x = np.linspace(-3,3,1000)
y = np.linspace(-3,3,1000)
X, Y = np.meshgrid(x, y)
Z1 = X * Y
fig = plt.figure(figsize = (15, 15))
ax = plt.axes(projection ='3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.plot_surface(X, Y, Z1, cmap='inferno', alpha=0.7)
z = np.linspace(0, 5, 1000)
x = np.linspace(0, 2, 1000)
X, Z = np.meshgrid(x, z)
Y = np.sqrt(X)
ax.plot_surface(X, Y, Z, color='orange')
ax.plot_surface(X, -Y, Z, color='orange')
z = np.linspace(0, 5, 1000)
y = np.linspace(0, 2, 1000)
Y, Z = np.meshgrid(y, z)
X = np.sqrt(Y)
ax.plot_surface(X, Y, Z, color='green')
ax.plot_surface(-X, Y, Z, color='green')
plt.show()
```

```python
u=np.linspace(-np.pi/2,np.pi/2,100)
v=np.linspace(0,2*np.pi,100)
U, V = np.meshgrid(u, v)
X = np.cos(U)*np.cos(V)
Y = 2*np.cos(U)*np.sin(V)
Z = np.sin(U)
fig=plt.figure(figsize =(10, 10))
ax = plt.axes(projection ='3d')
ax.plot_surface(X,Y,Z)
plt.show()
```
