```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
def method(x0,y0,ak,bk,n0,e):
    n1=2*n0
    n=0
    t=True
    a, b, c, d = [], [], [], [] #a - значение иксов при разбиении n, б - значение игриков при разбиении n и тд
    while True:
        a, b, c, d = [x0], [y0], [x0], [y0]
        dx1=(bk-ak)/(n0)
        dx2=(bk-ak)/(n1)
        maxl=0
        for i in range(1,n0+1):
            a.append(a[i-1]+dx1)
            b.append(b[i-1]+f(a[i-1],b[i-1])*dx1)
        for i in range(1,n1+1):
            c.append(c[i-1]+dx2)
            d.append(d[i-1]+f(c[i-1],d[i-1])*dx2)
        for i in range(1,n1):
            for j in range(1,n0):
                if c[i]==a[j]:
                    difference=abs(d[i]-b[j])
                    if difference>maxl:
                        maxl=difference
        if t:
            for i in range(1,n0+1):
                plt.plot([a[i-1],a[i]],[b[i-1],b[i]],"b-")
            for i in range(n1):
                plt.plot([c[i],c[i+1]],[d[i],d[i+1]],"k-")
            t=False
        if maxl<e:
            n=n1 
            for i in range(n):
                plt.plot([c[i],c[i+1]],[d[i],d[i+1]],"r-")
            break 
        else:
            n0=n1
            n1=2*n1
    plt.grid()
    print(n)
    return [c,d]
def f(x,y):
    return x*y
x=np.arange(0,1,0.001)
y_exact=np.exp((x**2)/2) #точное значение
plt.plot(x,y_exact,'-*')
mas=method(0,1,0,1,4,0.001)
```
