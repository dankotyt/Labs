```python
from sympy import Symbol
n=Symbol('n')
# q>1 
sp.pprint(sp.limit(2**n,n,np.inf))
# предел не равен 0 => необходимое условие сходимости не выполняется => расход 
# q=1 
sp.pprint(sp.limit(1**n,n,np.inf))
# предел не равен 0 => необходимое условие сходимости не выполняется => расход 
sp.pprint(sp.limit(0.5**n,n,np.inf)) 
#предел равен 0 => необходимо исследовать дальше, можно попытаться найти конечную сумму.
s=0.5
for i in range(2,100000):
    s+=0.5**i
print(s)
# предел частичных сумм равен 1 => ряд сходится. 
```
