```python
def bisection_method(a, b, epsilon):
    if f(a) * f(b) >= 0:
        print("На концах отрезка функция имеет одинаковый знак. Метод деления отрезка пополам не может быть применен.")
        return None

    iterations = 0
    while abs(b - a) > epsilon:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c
        iterations += 1
        
    return iterations
    

# Задаем начальные значения
a = -10
b = 10
epsilon = 0.0001

# Вызываем функцию для метода деления отрезка пополам
#print("Корень полинома: ", (a + b) / 2)
print("Число итераций: ", iterations)
bisection_method(a, b, epsilon)
```

```python
import numpy as np

def f(x):
    return x**3 - 3*x**2 - 9*x - 5

def f_prime(x):
    return 3*x**2 - 6*x - 9

def newton_method(x0, tol=1e-6, max_iter=100):
    x = x0
    iter_count = 0

    while abs(f(x)) > tol and iter_count < max_iter:
        x = x - f(x) / f_prime(x)
        iter_count += 1

    return x, iter_count

# Начальные значения для метода Ньютона
x0_values = [-10, -5, 0, 5, 10]

# Находим корни полинома и выводим результаты
for x0 in x0_values:
    root, iterations = newton_method(x0)
    print("Корень:", root)
    print("Число итераций:", iterations)
    print()
```

```python
#4
import numpy as np

def f(x):
    return x**3 - 3*x**2 - 9*x - 5

def f_prime(x):
    return 3*x**2 - 6*x - 9

def mod_newton_method(x0, m, tol=1e-6, max_iter=10000):
    x = x0
    iter_count = 0

    while abs(f(x)) > tol and iter_count < max_iter:
        x = x - m * f(x) / f_prime(x)
        iter_count += 1

    return x, iter_count

# Начальные значения для метода Ньютона
x0_values = [-10, -5, 0, 5, 10]
m = 2

# Находим корни полинома и выводим результаты
for x0 in x0_values:
    root, iterations = mod_newton_method(x0, m)
    print("Корень:", root)
    print("Число итераций:", iterations)
    print()
```

```python
#3
def square_func(a, epsilon = 1e-6, max_iter = 100):
    x = a / 2
    for i in range(max_iter):
        x_new = 0.5 * (a / x + x)
        if x - x_new < epsilon:
            break
        x = x_new

    return x

a = 16  # Число, из которого находим квадратный корень
sqrt = square_func(a)
print("Квадратный корень из", a, ":", sqrt)
```

```python
import numpy as np

def calculate_x(a):
    if a >= -1/4:
        # Если a >= -1/4, то уравнение имеет два действительных корня
        x1 = (-1 + np.sqrt(1 + 4 * a)) / 2
        x2 = (-1 - np.sqrt(1 + 4 * a)) / 2
        return x1, x2
    else:
        # Если a < -1/4, то уравнение имеет два комплексных корня
        real_part = -1 / 2
        imaginary_part = np.sqrt(-(4 * a + 1)) / 2
        x1 = complex(real_part, imaginary_part)
        x2 = complex(real_part, -imaginary_part)
        return x1, x2

# Пример использования
a = 2
x1, x2 = calculate_x(a)
print("x1 =", x1)
print("x2 =", x2)
```

```python
import math

# Графический метод
def graphical_method():
    import matplotlib.pyplot as plt
    import numpy as np

    x = np.linspace(-10, 10, 1000)  # Создаем массив значений x от -10 до 10
    y1 = np.sin(x)  # Функция sin x
    y2 = x / 2  # Функция x/2

    plt.plot(x, y1, label='sin(x)')
    plt.plot(x, y2, label='x/2')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black', linewidth=0.5)  # Ось x
    plt.axvline(0, color='black', linewidth=0.5)  # Ось y
    plt.grid(True, linestyle='--', alpha=0.7)  # Сетка
    plt.show()

# Метод половинного деления (бисекции)
def bisection_method():
    def f(x):
        return math.sin(x) - x / 2

    a = -10
    b = 10
    epsilon = 1e-6

    while abs(b - a) > epsilon:
        c = (a + b) / 2
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c

    return c

# Метод Ньютона
def newton_method():
    def f(x):
        return math.sin(x) - x / 2

    def f_prime(x):
        return math.cos(x) - 1 / 2

    x0 = 0  # Начальное приближение
    epsilon = 1e-6

    while True:
        x1 = x0 - f(x0) / f_prime(x0)
        if abs(x1 - x0) < epsilon:
            break
        x0 = x1

    return x1

# Вызываем все методы
graphical_method()
root_bisection = bisection_method()
root_newton = newton_method()

print("Корень, найденный методом половинного деления:", root_bisection)
print("Корень, найденный методом Ньютона:", root_newton)
```
