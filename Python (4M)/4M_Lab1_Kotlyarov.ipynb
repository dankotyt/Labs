```python
import math

def calculate_sin(x):
    term = x
    partial_sum = term
    n = 1

    while abs(term) > 1e-17:
        term *= -x * x / ((2 * n ) * (2 * n + 1))
        partial_sum += term
        n += 1
    
    return partial_sum

# Вычисление значения синуса в заданных точках
points = [0, math.pi/3, math.pi/2, math.pi, 2*math.pi]
for point in points:
    sin_value = calculate_sin(point)
    print(f"Значение синуса в точке {point}: {sin_value}")
    
print("==========================================")

# Вычисление значения синуса в точках 12π, 13π, 14π
more_points = [12*math.pi, 13*math.pi, 14*math.pi]
for point in more_points:
    term = point
    partial_sum = term
    n = 1
    
    print(f"Вычисление значения синуса в точке {point}:")
    print("Шаг\t\tПосчитанный член\t\tЧастичная сумма")
    
    while abs(term) > 1e-17:
        term *= -point * point / ((2 * n) * (2 * n + 1))
        partial_sum += term
        n += 1
        print(f"{n}\t\t{term}\t\t{partial_sum}")
```
