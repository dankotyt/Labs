```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, np.pi, 10)

y = np.sin(x)

plt.plot(x, y, 'ro') 
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('  f(x) = sin(x)   [0, ]')
plt.grid(True)
plt.show()

#  pi/9, ..   10 
h = np.pi / 9
dy = np.diff(y) / h

#       

#    ( )   . 
#             .
#    ( )     . 
#                 .
coefficients = np.polyfit(x, y, 9)

#  
derivative_coefficients = np.polyder(coefficients)

print(' :')
print(coefficients)
print('   :')
print(derivative_coefficients)
 
```

```python
import numpy as np
import matplotlib.pyplot as plt

#   
x = np.linspace(0, np.pi, 100)
f_x = np.sin(x)

#      
x_points = [0, np.pi/6, np.pi/4, np.pi/3, np.pi/2]
f_prime_numerical = np.gradient(f_x, x)

#     
f_prime_analytical = np.cos(x)

#      
for i in range(len(x_points)):
    print(f"    x = {x_points[i]}: {f_prime_numerical[np.abs(x - x_points[i]).argmin()]}")
    print(f"    x = {x_points[i]}: {f_prime_analytical[np.abs(x - x_points[i]).argmin()]}\n")

#   
error = np.abs(f_prime_numerical - f_prime_analytical)

#    
for i in range(len(x_points)):
    print(f"    x = {x_points[i]}: {error[np.abs(x - x_points[i]).argmin()]}\n")

#  
plt.plot(x, f_prime_numerical, label=' ')
plt.plot(x, f_prime_analytical, label=' ')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.legend()
plt.title('     ')
plt.show()


```
