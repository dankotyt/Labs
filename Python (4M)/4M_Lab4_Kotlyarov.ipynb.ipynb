```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, np.pi, 10)

y = np.sin(x)

plt.plot(x, y, 'ro') 
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Значения функции f(x) = sin(x) на отрезке [0, π]')
plt.grid(True)
plt.show()

#Зададим шаг pi/9, т.к. разбиение на 10 частей
h = np.pi / 9
dy = np.diff(y) / h

# Интерполяция полиномом с использованием метода конечных разностей

#Вычисляются разности первого порядка (первые разности) между соседними точками. 
#Для этого вычитается значение функции в предыдущей точке из значения функции в текущей точке.
#Вычисляются разности второго порядка (вторые разности) между соседними разностями первого порядка. 
#Для этого вычитается значение разности первого порядка в предыдущей точке из значения разности первого порядка в текущей точке.
coefficients = np.polyfit(x, y, 9)

# Дифференцирование полинома
derivative_coefficients = np.polyder(coefficients)

print('Коэффициенты полинома:')
print(coefficients)
print('Коэффициенты полинома при дифференцировании:')
print(derivative_coefficients)
 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Задаем значения точек
x = np.linspace(0, np.pi, 100)
f_x = np.sin(x)

# Находим численные производные в заданных точках
x_points = [0, np.pi/6, np.pi/4, np.pi/3, np.pi/2]
f_prime_numerical = np.gradient(f_x, x)

# Вычисляем точные аналитические значения производной
f_prime_analytical = np.cos(x)

# Выводим значения численных и аналитических производных
for i in range(len(x_points)):
    print(f"Численная производная в точке x = {x_points[i]}: {f_prime_numerical[np.abs(x - x_points[i]).argmin()]}")
    print(f"Аналитическая производная в точке x = {x_points[i]}: {f_prime_analytical[np.abs(x - x_points[i]).argmin()]}\n")

# Вычисляем практическую ошибку
error = np.abs(f_prime_numerical - f_prime_analytical)

# Выводим значения практической ошибки
for i in range(len(x_points)):
    print(f"Практическая ошибка в точке x = {x_points[i]}: {error[np.abs(x - x_points[i]).argmin()]}\n")

# Строим графики
plt.plot(x, f_prime_numerical, label='Численное решение')
plt.plot(x, f_prime_analytical, label='Аналитическое решение')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.legend()
plt.title('Сравнение численного и аналитического решений производной')
plt.show()


```
